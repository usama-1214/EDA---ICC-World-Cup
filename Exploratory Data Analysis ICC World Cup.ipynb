```python
import os
# Define the folder path where your cleaned CSV files are stored
folder_path = r'D:\Bootcamp\Python\ICC Cricket World Cup\Cleaned'

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize an empty dictionary to store DataFrames by file names
data_frames = {}

# Read each CSV file and store DataFrames by file names in the dictionary
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    data_frames[file] = df

# Combine DataFrames in the dictionary into a single DataFrame
combined_df = pd.concat(data_frames.values(), axis=1)
```

```python
pd.set_option('display.max_columns', None)

combined_df.head()
```

# Statistical Analysis

## Mean, Median and Mode

```python
import pandas as pd

# Calculate mean, median, mode, and standard deviation for runs and wickets for each tournament
# Group by tournament and calculate mean, median, and standard deviation for runs and wickets
statistics_by_tournament = combined_df.groupby('Team_Name').agg({
    'runs': ['mean', 'median', 'std'],
    'wkts': ['mean', 'median', 'std']
}).reset_index()

# Rename the columns for better readability
statistics_by_tournament.columns = ['Tournament', 
                                     'Mean Runs', 'Median Runs', 'Std Dev Runs',
                                     'Mean Wickets', 'Median Wickets', 'Std Dev Wickets']

# Display the results
print(statistics_by_tournament)
#statistics_by_tournament.info()
```

# Visualizations

## Total Runs Scored By Teams


```python
plt.figure(figsize=(12, 8))
team_runs = combined_df.groupby('abbreviation')['runs'].sum().sort_values(ascending=False)
team_runs.plot(kind='bar', color='skyblue')
plt.title('Total Runs Scored by Teams')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='runs', y='wkts', data=combined_df, hue='abbreviation', palette='viridis')
plt.title('Relationship between Runs and Wickets')
plt.xlabel('Runs')
plt.ylabel('Wickets')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python

# Group by venue and calculate the total runs scored at each country
venue_runs = combined_df.groupby('country')['runs'].sum().reset_index()

# Sort the data by total runs in descending order
venue_runs = venue_runs.sort_values(by='runs', ascending=False)

# Plot the bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='runs', y='country', data=venue_runs, palette='viridis')
plt.title('Total Runs Scored at Each Country')
plt.xlabel('Total Runs')
plt.ylabel('Venue')
plt.show()
```
